```python
colors = {'setosa':'red', 'versicolor':'green', 'virginica':'blue'}

plt.figure()
fig, myaxis = plt.subplots(1,1)
for species in df.species.unique():
    df[df.species == species].plot(ax=myaxis, x='sepal_length', y='sepal_width', 
                                   kind='scatter', label=species,
                                  color=colors[species])

plt.show()
```

```python
plt.figure()
fig, myaxis = plt.subplots(1,1)
df[df.species == 'setosa'].plot(ax=myaxis, x='sepal_length', y='sepal_width', kind='scatter', label='setosa', color='red')
df[df.species == 'virginica'].plot(ax=myaxis, x='sepal_length', y='sepal_width', kind='scatter', label='virginica', color='green')
plt.show()
```

```python
from itertools import combinations

cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for comb in combinations(cols, 2):
    plt.figure()
    fig, myaxis = plt.subplots(1,1)
    for species in df.species.unique():
        df[df.species == species].plot(ax=myaxis, x=comb[0], y=comb[1], 
                                       kind='scatter', label=species,
                                      color=colors[species])

    plt.show()
```
